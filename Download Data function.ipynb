```python
import os
import json
import pandas as pd
import seaborn as sns
from datetime import datetime
import requests
```

```python
data = requests.get('https://ckan0.cf.opendata.inter.prod-toronto.ca/dataset/0b6d3a00-7de1-440b-b47c-7252fd13929f/resource/355d4464-eb3c-4780-af79-43dd533ae906/download/Motor%20Vehicle%20Collisions%20with%20KSI%20Data.geojson').json()
```

```python
# Initialize an empty list to store flattened dictionaries
flat_data = []

# Iterate through the 'features' list in the JSON data
for feature in data['features']:
    properties = feature['properties']
    geometry = feature['geometry']
    flat_properties = properties.copy()
    flat_properties.update(geometry)
    flat_data.append(flat_properties)
    
# Create a DataFrame from the flattened data
df = pd.DataFrame(flat_data)
```
